# Задание 3. Задача с качеством уроков

```python
import pandas as pd

# Читаем все файлы, которые будут необходимы для выполнения данного задания

lessons_data = pd.read_csv('/Users/garynich/Documents/Tetrica School/tech_quality/lessons.txt', sep='|')
lessons_data = lessons_data[1:] #Удаляем первую строку, так как она несет в себе мусор -  особенность чтения  txt
lessons_data = lessons_data[:-1]

participants_data = pd.read_csv('/Users/garynich/Documents/Tetrica School/tech_quality/participants.txt', sep='|')
participants_data = participants_data[1:] #Удаляем первую строку, так как она несет в себе мусор -  особенность чтения  txt
participants_data = participants_data[:-1]

quality_data = pd.read_csv('/Users/garynich/Documents/Tetrica School/tech_quality/quality.txt', sep='|')
quality_data = quality_data[1:] #Удаляем первую строку, так как она несет в себе мусор -  особенность чтения  txt
quality_data = quality_data[:-1]

users_data = pd.read_csv('/Users/garynich/Documents/Tetrica School/tech_quality/users.txt', sep='|')
users_data = users_data[1:] #Удаляем первую строку, так как она несет в себе мусор -  особенность чтения  txt
users_data = users_data[:-1]

```

```python
lessons_data.columns = ['id', 'event_id', 'subject', 'scheduled_time']

lessons_data['event_id'] = lessons_data['event_id'].astype(int) #Переводим id в целочисленный тип для упрощения работы
lessons_data['scheduled_time'] =  pd.to_datetime(lessons_data['scheduled_time'])
lessons_data['mow_time'] = lessons_data['scheduled_time'] + pd.to_timedelta(3, unit='h') #Создаем столбец 
            #с проведением урока по московскому времени
lessons_data['subject'] = lessons_data['subject'].str.strip() # Удаляем пробелы со всех сторон у значения


for row in lessons_data.columns:
    print('Количество пропусков в столбце ', row, '=', lessons_data[row].isnull().sum())
print(lessons_data.info())

lessons_data.head(3)
```

```python
participants_data.columns = ['event_id', 'user_id']
participants_data['user_id'] = participants_data['user_id'].astype(str)
participants_data['event_id'] = participants_data['event_id'].astype(int)
for row in participants_data.columns:
    print('Количество пропусков в столбце ', row, '=', participants_data[row].isnull().sum())
print(participants_data.info())


participants_data.head(3)
```

```python
quality_data.columns = ['lesson_id', 'tech_quality']
quality_data['tech_quality'] = quality_data['tech_quality'].str.strip() # Удаляем пробелы со всех сторон у значения
quality_data['tech_quality'] = pd.to_numeric(quality_data['tech_quality'])

for row in quality_data.columns:
    print('Количество пропусков в столбце ', row, '=', quality_data[row].isnull().sum())
print(quality_data.info())

quality_data.head(3)
```

```python
users_data.columns = ['user_id', 'role']
users_data['user_id'] = users_data['user_id'].astype(str)
for row in users_data.columns:
    print('Количество пропусков в столбце ', row, '=', users_data[row].isnull().sum())
print(users_data.info())

users_data.head(3)
```

Выполнена предобработка предоставленных файлов. Данные переведены в тип, который пригоден для дальнейшей работы. 

##  1. Найти все уроки по физике (subject=phys).

```python
lessons_data  = lessons_data.query('subject == "phys"')
lessons_data.head(5)
```

## 2. В каждый день для каждого репетитора посчитать среднюю арифметическую оценку за его уроки (учитывать только уроки из п. 1).

```python
rep = quality_data.pivot_table(index = 'lesson_id', values = 'tech_quality', aggfunc = ['count', 'sum'])
rep.columns = ['count_quality', 'sum_quality']
rep.reset_index(inplace = True)
rep.head(5)
```

```python
lessons_datas = pd.merge(lessons_data, rep, left_on='id', right_on='lesson_id')

lessons_datas['mow_day'] = lessons_datas['mow_time'].dt.date
lessons_datas['mean_quality'] = lessons_datas['sum_quality'] /  lessons_datas['count_quality']
lessons_datas.drop(['scheduled_time', 'lesson_id', 'mow_time'], axis=1, inplace=True)


part_with_roles = participants_data.join(users_data, lsuffix='_l')
part_with_roles.drop(['user_id_l'], axis=1, inplace=True)

teachers_with_event = part_with_roles.query('role == " tutor"')

finish_data  = pd.merge(lessons_datas, teachers_with_event, on = 'event_id')
finish_data.head(5)
```

## 3. Найти учителя, который в этот день имеет самую низкую среднюю арифметическую оценку за уроки 

```python
pre = finish_data.groupby(['mow_day', 'user_id']).agg({'mean_quality':'mean'}).round(2).reset_index()
report = pre.pivot_table(index = 'mow_day', aggfunc = 'min')

report[['user_id', 'mean_quality']].reset_index()
```
